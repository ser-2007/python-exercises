### You can loops when reading files

```python
sea = open("fishes.txt", "r")
for line in sea:
    print(line)
sea.close()
```

```python
sea = open("fishes.txt", "r")
for line in sea.readlines():
    print(line)
sea.close()
```

```python
rumi = open("C:\\Users\\serkan\\Desktop\\rumi.txt", "r")
for line in rumi.readlines():
    print(line)
rumi.close()
```

```python
rumi = open("C:\\Users\\serkan\\Desktop\\rumi.txt", "r")
for line in rumi:
    print(line)
rumi.close()
```

```python
with open ("fishes.txt", "r") as sea: # open ve close birlikte yuapilabiliyor girintinin icine istediginiizi yazabiliyorsunuz
    print(sea.read())
```

### Writing Files

```python
with open("dummy_file.txt", "w", encoding = "utf-8") as file:
    file.write("This is the new line for my dummy_file")
```

```python
with open("dummy_file.txt", "r", encoding = "utf-8") as file:
    print(file.read())
```

```python
with open("dummy_file.txt", "w", encoding = "utf-8") as file:
    file.write("This is the new line for my dummy_file, but it is different.\n")
```

```python
with open("dummy_file.txt", "w", encoding = "utf-8") as file:
    file.write("This is the new line for my dummy_file.\n but it is not new . \nThis is the last")
```

```python
with open("dummy_file.txt", "w", encoding = "utf-8") as file:
    file.write("This is my first sentence")
    file.write("This is my second sentence")
    file.write("This is my thirth sentence\n")
    file.write("This is my fourth sentence")
    file.write("This is my fifth sentence")
with open("dummy_file.txt", "r", encoding = "utf-8") as file:
    print(file.read())
    
```

```python
with open("fruits.txt", "w", encoding = "utf-8") as file:
    file.write("Banana\n")
    file.write("Orange\n")
    file.write("Apple\n")
    file.write("Strawberry\n")
    file.write("Cherry\n")
with open("fruits.txt", "r", encoding = "utf-8") as file:
    print(file.read())
    
```

```python
with open("fruit.txt","w",encoding="utf- 8") as file:
    fruits =["Banana", "Orange", "Apple", "Strawberry", "Cherry"]
    for i in fruits:
        file.write(i+"\n")
with open("fruit.txt", "r", encoding = "utf-8") as file:
    print(file.read())
```
